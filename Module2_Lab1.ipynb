2) Do you think maximum/minimum values are a clear indication of which body-type is costliest, or does the interpretation improve by also taking the interquartile range into consideration?

```python
sns.boxplot(x='body-style', y=y, data=data)
plt.xlabel("Body Type")
plt.ylabel("Price")
plt.title("Price Distribution by Body Type")
plt.show()
```

While maximum values provide an insight into the most expensive cars in each body type, they may not always be representative due to potential outliers.
Taking the interquartile range (IQR) into consideration is important because it provides a measure of the spread of prices within each body type.
IQR considers the middle 50% of the data, excluding outliers. A body type with a high maximum but a small IQR may have only a few very expensive outliers but generally lower-priced cars.

## 3.Do you think the body-style of a car can be a good predictor for its price?

The body style of a car can provide some information about its price, but it may not be the sole or most accurate predictor.
The plot can indicate how prices vary across different body types, but other factors like engine size, brand, features, and market demand also play significant roles in determining a car's price.
To make a more accurate prediction of a car's price, a combination of multiple features and machine learning models would be necessary.

```python
mean_prices = data.groupby('body-style')['price'].mean()

sns.barplot(x=mean_prices.index, y=mean_prices.values)
plt.xlabel("Body Style")
plt.ylabel("Mean Price")
plt.title("Mean Price vs Body Type")
plt.xticks(rotation=45)
plt.show()

```
