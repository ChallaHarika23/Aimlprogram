```python
import numpy as np
from numpy import linalg
from numpy.linalg import norm
from scipy.spatial.distance import squareform, pdist

import sklearn
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
from sklearn.preprocessing import scale

RS = 31060120

import matplotlib.pyplot as plt
import matplotlib.patheffects as PathEffects
import matplotlib
%matplotlib inline

import seaborn as sns
sns.set_style('darkgrid')
sns.set_palette('muted')
sns.set_context("notebook", font_scale=1.5,
                rc={"lines.linewidth": 2.5})
```

```python
dig = load_digits()
dig.data.shape
```

```python
nrows, ncols = 3, 6
plt.figure(figsize=(7,4))
plt.gray()
for i in range(ncols * nrows):
    ax = plt.subplot(nrows, ncols, i + 1)
    ax.matshow(dig.images[i,...])
    plt.xticks([]); plt.yticks([])
    plt.title(dig.target[i])
```

```python
X = np.vstack([dig.data[dig.target==i]
               for i in range(10)])
y = np.hstack([dig.target[dig.target==i]
               for i in range(10)])
```

```python
dig_pr = TSNE(init="pca", random_state=RS).fit_transform(X)
```

```python
def scatter(x, colors):
    palette = np.array(sns.color_palette("hls", 10))

    f = plt.figure(figsize=(10, 10))
    ax = plt.subplot(aspect='equal')
    sc = ax.scatter(x[:,0], x[:,1], lw=0, s=30,
                    c=palette[colors.astype(int)])
    plt.xlim(-30, 30)
    plt.ylim(-30, 30)
    ax.axis('off')
    ax.axis('tight')

    txts = []
    for i in range(15):
        xtext, ytext = np.median(x[colors == i, :], axis=0)
        txt = ax.text(xtext, ytext, str(i), fontsize=22)
        txt.set_path_effects([
            PathEffects.Stroke(linewidth=7, foreground="w"),
            PathEffects.Normal()])
        txts.append(txt)
    plt.show()

    return f, ax, sc

scatter(dig_pr, y)
```

# QUESTION:
Now that you understand t-SNE a bit better, can you point out some differences between PCA and t-SNE. What are the advantages/disadvantages of one over the other?

Advantages of PCA:

1.Linear and interpretable transformation.
2.Useful for reducing dimensionality when you want to maintain the global structure of the data.

Disadvantage of PCA:

Limited in capturing complex, non-linear relationships in the data.

Advantages of t-SNE:

1.Effective at capturing non-linear patterns.
2.Well-suited for visualizing clusters and revealing hidden patterns in the data.
3.provides more accurate representation of high-dimensional data.

Disadvantages of t-SNE:

1.More expensive, especially with a large dataset.
2.Sensitive to hyperparameters, making parameter tuning important.


# 1.Which of the two algorithms is linear and which one is non-linear?

PCA is a linear technique, while t-SNE is a non-linear technique. PCA finds orthogonal axes in the original feature space which effectively reducing dimensionality by projecting data points onto these linear axes. t-SNE, on the other hand, tries to preserve the pairwise similarities between data points, which makes it more effective at capturing non-linear relationships in the data.




# 2.How does the non-linearity in one of these two algorithms help in capturing certain data sets?


The non-linearity in t-SNE helps capture complex, non-linear relationships in high-dimensional data. It preserves local structures, making it effective for identifying clusters and subgroups. It handles data distributed on non-linear manifolds and separates signal from noise. t-SNE is valuable for pattern discovery and revealing hidden structures in data.

# 3.PCA is known to keep points which were further apart in the higher dimension, far apart in the lower dimension as well. Does t-SNE do the same? Or does it try to preserve local neighbourhood?


No,t-SNE doesn't do the same as PCA. t-SNE primarily focuses on preserving local neighborhoods and pairwise similarities. It attempts to keep similar data points close together and push dissimilar points apart

# 4.Can you comment on which one of the two is computationally more expensive?

t-SNE is generally computationally more expensive than PCA.Tuning t-SNE's hyperparameters, such as the perplexity and the number of iterations, can also affect its computational cost.
