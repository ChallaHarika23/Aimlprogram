```python
def tuple2string(tup):
  st = ''
  for ii in tup:
    st = st + ii
  return st

def key2string(keys):
  return [tuple2string(i) for i in keys]

# plot the histogram
def plothistogram(ngram):
  keys = key2string(ngram.keys())
  values = list(ngram.values())

  combined = zip(keys, values)
  zipped_sorted = sorted(combined, key=lambda x: x[0])
  keys, values = map(list, zip(*zipped_sorted))
  plt.bar(keys, values)
```

```python
unigram_eng1 = Counter(ngrams(e1,1))
plothistogram(unigram_eng1)
plt.title('1')
plt.show()
unigram_eng2 = Counter(ngrams(e2,1))
plothistogram(unigram_eng2)
plt.title('2')
plt.show()
```

```python
def plotbihistogram(ngram):
  freq = np.zeros((26,26))
  for ii in range(26):
    for jj in range(26):
      freq[ii,jj] = ngram[(chr(ord('a')+ii), chr(ord('a')+jj))]
  plt.imshow(freq, cmap = 'jet')
  return freq
```

```python
bieng1 = plotbihistogram(bigram_eng1)
plt.show()
bieng2 = plotbihistogram(bigram_eng2)
```

```python
bifr1 = plotbihistogram(bigram_fr1)
plt.show()
bifr2 = plotbihistogram(bigram_fr2)
```

```python
from IPython.core.debugger import set_trace

def ind2tup(ind):
  ind = int(ind)
  i = int(ind/26)
  j = int(ind%26)
  return (chr(ord('a')+i), chr(ord('a')+j))

def ShowTopN(bifreq, n=10):
  f = bifreq.flatten()
  arg = np.argsort(-f)
  for ii in range(n):
    print(f'{ind2tup(arg[ii])} : {f[arg[ii]]}')
```

```python
print('\nEng 1:')
ShowTopN(bieng1)
print('\nEng 2:')
ShowTopN(bieng2)
print('\nFren 1:')
ShowTopN(bifr1)
print('\nFren 2:')
ShowTopN(bifr2)
```

```python
def cumArray(img):
  img2 = img.copy()
  for ii in range(1, img2.shape[1]):
    img2[ii,:] = img2[ii,:] + img2[ii-1,:]
  img2 = img2>0
  return img2

def getHolePixels(img):
  im1 = cumArray(img)
  im2 = np.rot90(cumArray(np.rot90(img)), 3)
  im3 = np.rot90(cumArray(np.rot90(img, 2)), 2)
  im4 = np.rot90(cumArray(np.rot90(img, 3)), 1)
  hull =  im1 & im2 & im3 & im4
  hole = hull & ~ (img>0)
  return hole
```

```python
imgs = [num7[456,:,:],  num6[456,:,:]]
for img in imgs:
  plt.subplot(1,2,1)
  plt.imshow(getHolePixels(img))
  plt.subplot(1,2,2)
  plt.imshow(img)
  plt.show()
```

```python
hole1 = np.array([getHolePixels(i).sum() for i in num7])
hole0 = np.array([getHolePixels(i).sum() for i in num6])

plt.hist(hole1, alpha=0.7);
plt.hist(hole0, alpha=0.7);
```

```python
def getHullPixels(img):
  im1 = cumArray(img)
  im2 = np.rot90(cumArray(np.rot90(img)), 3)
  im3 = np.rot90(cumArray(np.rot90(img, 2)), 2)
  im4 = np.rot90(cumArray(np.rot90(img, 3)), 1)
  hull =  im1 & im2 & im3 & im4
  return hull
```

```python
imgs = [num7[456,:,:],  num6[456,:,:]]
for img in imgs:
  plt.subplot(1,2,1)
  plt.imshow(getHullPixels(img))
  plt.subplot(1,2,2)
  plt.imshow(img)
  plt.show()
```

```python
hull1 = np.array([getHullPixels(i).sum() for i in num7])
hull0 = np.array([getHullPixels(i).sum() for i in num6])

plt.hist(hull1, alpha=0.7);
plt.hist(hull0, alpha=0.7);
```

```python
def minus(a, b):
  return a & ~ b

def getBoundaryPixels(img):
  img = img.copy()>0
  rshift = np.roll(img, 1, 1)
  lshift = np.roll(img, -1 ,1)
  ushift = np.roll(img, -1, 0)
  dshift = np.roll(img, 1, 0)
  boundary = minus(img, rshift) | minus(img, lshift) | minus(img, ushift) | minus(img, dshift)
  return boundary
```

```python
imgs = [num7[456,:,:],  num6[456,:,:]]
for img in imgs:
  plt.subplot(1,2,1)
  plt.imshow(getBoundaryPixels(img))
  plt.subplot(1,2,2)
  plt.imshow(img)
  plt.show()
```

```python
bound1 = np.array([getBoundaryPixels(i).sum() for i in num7])
bound0= np.array([getBoundaryPixels(i).sum() for i in num6])

plt.hist(bound1, alpha=0.7);
plt.hist(bound0, alpha=0.7);
```

```python
def calculateSymmetry(img):
    if img.shape[1] % 2 == 0:
        left_half = img[:, :img.shape[1] // 2]
        right_half = img[:, img.shape[1] // 2:]
        symmetry_diff = np.abs(left_half - np.flip(right_half, axis=1))


        symmetry_score = np.sum(symmetry_diff)


        symmetry_score /= img.size

        return symmetry_score
    else:

        return 0.0

imgs = [num7[456, :, :], num6[456, :, :]]

for img in imgs:
    plt.subplot(1, 2, 1)
    plt.imshow(getHullPixels(img))
    plt.subplot(1, 2, 2)
    plt.imshow(img)
    plt.title(f"Symmetry Score: {calculateSymmetry(img):.2f}")
    plt.show()


symmetry1 = np.array([calculateSymmetry(i) for i in num7])
symmetry0 = np.array([calculateSymmetry(i) for i in num6])


```

```python
plt.hist(symmetry1, alpha=0.7, label='Class 1')
plt.hist(symmetry0, alpha=0.7, label='Class 0')
plt.xlabel('Symmetry Score')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
